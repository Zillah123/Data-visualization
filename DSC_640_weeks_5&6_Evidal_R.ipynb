```r
library(ggplot2)
library(readxl)
library(treemapify)
```

```r
my_data<-read.csv("F:\\Data\\unemployement-rate-1948-2010.csv")
str(my_data)
```

```r
my_data
```

```r
ggplot(my_data, aes(area = Value, fill=Period, label=Period))+geom_treemap()
```

```r
ggplot(my_data, aes(area = Value, fill=Period, subgroup=Period))+geom_treemap()
```

```r
treemap_coords <- treemapify(my_data, area="Value", fill="Period", label="Year", group="Period")
head(treemap_coords)
```

```r
ggplot(my_data, aes(area = Value, fill=Year, subgroup=Year))+geom_treemap()
```

```r
ggplot(my_data, aes(x=Year, y=Value))+geom_area()
```

```r
ggplot(my_data, aes(x=Year, y=Value))+geom_area(fill='#142F86', alpha=10)+geom_line(color='skyblue', alpha=10)
```

```r
p <- ggplot(my_data, aes(x=Year))
p+geom_area(stat="bin")
```

```r
ggplot(my_data, aes(x=Year, y=Value, fill=Period))+geom_area()+ggtitle("Stacked Area chart of Value, Year, and Period")
```
