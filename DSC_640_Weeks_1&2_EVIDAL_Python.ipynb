```python
for c in df.columns:
    miss = df[c].isnull().sum()
    if miss>0:
        print("{} has {} missing value(s)".format(c,miss))
    else:
        print("{} has NO missing value!".format(c))
```

```python
fig, ax = plt.subplots()
ax.bar(df["Issue"], df["Approve"])
ax.set_xticklabels(df["Issue"], rotation=90)
ax.set_ylabel("Approval Ratings")
ax.set_xlabel("Issue")
ax.set_title("Obama's Approval Ratings")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(df["Issue"],df["Approve"], label="Approval")
ax.bar(df["Issue"],df["Disapprove"], bottom=df["Approve"], label="Disapproval")
ax.set_xticklabels(df["Issue"], rotation=90)
ax.set_ylabel("Ratings")
ax.set_xlabel("Issue")
ax.set_title("Obama's Ratings")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(df["Issue"],df["Approve"], label="Approval")
ax.bar(df["Issue"],df["Disapprove"], bottom=df["Approve"], label="Disapproval")
ax.bar(df["Issue"],df["None"], label="None")
ax.set_xticklabels(df["Issue"], rotation=90)
ax.set_ylabel("Ratings")
ax.set_xlabel("Issue")
ax.set_title("Obama's Ratings")
ax.legend()
plt.show()
```

```python
race_relations = [52, 38, 10]
labels = 'Approve','Disapprove','None'
explode = (0, 0.1, 0) 
fig, ax = plt.subplots()
ax.pie(race_relations, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')
ax.set_title("Obama's Ratings for Race Relations")
ax.legend(bbox_to_anchor=(1,1), loc="upper left",bbox_transform=plt.gcf().transFigure)
plt.show()
```

```python
approve=df["Approve"]
issue=df["Issue"]
fig, ax = plt.subplots()
ax.pie(approve,labels=issue, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')
ax.set_title("Percentage of issues that make up Obama's Approval ratings")
ax.legend(bbox_to_anchor=(1.5,0), loc="lower right",bbox_transform=plt.gcf().transFigure)
plt.show()
```

```python
race_relations = [52, 38, 10]
labels = 'Approve','Disapprove','None'
colors = ['red','blue','green']
explode = (0, 0, 0) 
plt.pie(race_relations, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
centre_circle = plt.Circle((0,0),0.75,color='black',fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.title("Donut Chart of Obama's ratings for Race Relations")
plt.show()
```
